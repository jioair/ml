```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
import matplotlib.pyplot as plt
iris = load_iris()
df = pd.DataFrame(iris.data,columns=iris.feature_names)
print(iris.data)
df.head(10)
```

```python
print(df)
```

```python
df['target'] = iris.target
df['flower_name'] = df.target.apply(lambda x:iris.target_names[x])
df.head(10)
```

```python
for x in range(3):
    print(iris.target_names[x])
```

```python
df_1 = df[df["target"]==0][:40]
df_2 = df[df["target"]==1][:10]
df_3 = df[df["target"]==2][:10]
final_df = pd.concat([df_1,df_2,df_3])
print(final_df)
```

```python
map_characters = {0: 'Setosa', 1: 'versicolor', 2: 'virginica'}
dict_characters = map_characters
import seaborn as sns
lab = final_df['target']
dist = lab.value_counts()
sns.countplot(lab)
```

```python
print(lab)
```

```python
print(dist)
```

```python
from sklearn.neighbors import KNeighborsClassifier
x = final_df.iloc[:,:4].values
y = final_df.iloc[:,4].values
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size = 0.2 )
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train,Y_train)
y_pred = knn.predict(X_test)
knn.score(X_test,Y_test)
```

```python
from sklearn.metrics import classification_report
print (classification_report(Y_test,y_pred))
```

```python
from imblearn.over_sampling import RandomOverSampler, SMOTE, ADASYN
from imblearn.under_sampling import RandomUnderSampler, NearMiss, TomekLinks
from collections import Counter
smote = SMOTE()
X_smote_resampled,Y_smote_resampled = smote.fit_resample(X_train,Y_train)
print("SMOTE")
print("Before Resampling: ",sorted(Counter(Y_train).items()))
print("After Resampling: ", sorted(Counter(Y_smote_resampled).items()))
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_smote_resampled,Y_smote_resampled)
y_pred = knn.predict(X_test)
from sklearn.metrics import classification_report
print(knn.score(X_test,Y_test))
print(classification_report(Y_test,y_pred))
```

```python
rus = RandomUnderSampler()
X_rus_resampled,Y_rus_resampled = rus.fit_resample(X_train,Y_train)
print("Random Under Sampling")
print("Before Resampling: ",sorted(Counter(Y_train).items()))
print("After Resampling: ", sorted(Counter(Y_rus_resampled).items()))
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_rus_resampled,Y_rus_resampled)
y_pred = knn.predict(X_test)
from sklearn.metrics import classification_report
print(knn.score(X_test,Y_test))
print(classification_report(Y_test,y_pred))
```
