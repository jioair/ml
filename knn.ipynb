```python
#KNN Method-2

import pandas as pd
import math
your_dataset = pd.read_csv('knn2.csv')
print(your_dataset)

a=  your_dataset['temperature'].values
b=  your_dataset['outlook'].values
c=  your_dataset['humidity'].values
d=  your_dataset['wind'].values
final2D = []

for i in range(len(a)):
    insideData = []
    insideData.append(a[i])
    insideData.append(b[i])
    insideData.append(c[i])
    insideData.append(d[i])
    final2D.append(insideData)

print(final2D)

instance = [3, 1,2,2]

calculatedList = []

for i in range(len(a)):
    s = 0
    
    for j in range(len(instance)):
        if final2D[i][j] == instance[j]:
            s += instance[j] * 1
        else:
            s += 0
    calculatedList.append(s/4)
print(calculatedList)

calMain = calculatedList
calculatedList = set(calculatedList)
calculatedList = list(calculatedList)
calculatedList.sort()
k=3
instanceSol = []
for i in range(k):
    instanceSol.append(calculatedList[i])
print(instanceSol)

di = {}

for i in range(len(calMain)):
    if calMain[i] in di:
        di[calMain[i]] += 1
    else:
        di[calMain[i]] = 1

print('dic', di)
sorted(di)
print(di)

c = 0  #count for number of NO's
d = 0 #count for number of Yes's
for i in range(len(instanceSol)):
    if instanceSol[i] > 0.5:
        d += di[instanceSol[i]]
    elif instanceSol[i] < 0.5:
        c += di[instanceSol[i]]
    else:
        d += di[instanceSol[i]]
        c += di[instanceSol[i]]
print( c,d)

if d>c:
    print('Yes,you can play game')
elif d<c:
    print('No you cant play game')
```
